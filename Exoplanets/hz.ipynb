```python
import numpy as np
import matplotlib.pyplot as plt


def boundary(mstar, albedo, temp, nu):
    "Calculate habitable zone around a star"
    sigma=5.67e-5 
    Lsun=3.9e33
    
    return np.sqrt(Lsun*(1-albedo)/(16*np.pi*sigma))*np.power(mstar, nu/2.)*np.power(1/temp,2)
    
albedo=0.2
au=1.496e13    

def MLnu(M_star):
    "Return the mass luminosity relation exponent"
    if (M_star<=1):
        return(2.5)
    if (M_star>1 and M_star<10):
        return(4.5)
    if (M_star>=10):
        return(3)

mstar=np.arange(0.08,40,0.01)

nu=np.arange(0.08,40,0.01)

i=0
for star in mstar:  
    nu[i]=MLnu(star)
    i=i+1


amin=boundary(mstar, albedo, 373,nu)/au
amax=boundary(mstar, albedo, 273,nu)/au

plt.xscale('log')
plt.yscale('linear')
plt.xlabel('Orbital Radius (AU)')
plt.ylabel('Stellar Mass (Msun)')

plt.xlim (1e-1, 100)
plt.ylim (0.5e-1, 2)


plt.plot(amin, mstar, color='black')
plt.plot(amax, mstar)
plt.scatter(1,1,label='Earth')
plt.legend(loc="upper left",fontsize=13)

plt.show()
```
