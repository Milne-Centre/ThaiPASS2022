```python
import numpy as np
import matplotlib.pyplot as plt


def boundary(mstar, albedo, temp):
    "Calculate habitable zone around a star"
    sigma=5.67e-5 
    Lsun=3.9e33
    
    return np.sqrt(Lsun*(1-albedo)/(16*np.pi*sigma))*np.power(mstar, 3/2.)*np.power(1/temp,2)
    
albedo=0.2
au=1.496e13    

mstar=np.arange(0.08,2,0.01)

amin=boundary(mstar, albedo, 373)/au
amax=boundary(mstar, albedo, 273)/au

plt.xscale('log')
plt.yscale('linear')
plt.xlabel('Orbital Radius (AU)')
plt.ylabel('Stellar Mass (Msun)')

plt.xlim (1e-2, 10)
plt.ylim (0, 2)


plt.plot(amin, mstar, color='black')
plt.plot(amax, mstar)
plt.scatter(1,1,label='Earth')
plt.legend(loc="upper left",fontsize=13)

plt.show()
```
